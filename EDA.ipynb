```python
#CROSSTAB

crosstab_dir_viento_estaciones = pd.crosstab(df_ozono['nombre_estacion'], df_ozono['dir_cardinal'])
crosstab_dir_viento_estaciones
```

```python
#STACKED

plt.style.use('tableau-colorblind10')

crosstab_dir_viento_estaciones.plot(
    kind='bar', 
    stacked=True, 
    figsize=(10, 6), 
    color=sns.color_palette("mako", n_colors=len(crosstab_dir_viento_estaciones.columns))  
)

plt.title('Distribución de Direcciones Cardinales por Estación', fontsize=14, fontweight='bold', color='black')
plt.xlabel('Nombre de la Estación', fontsize=12, fontweight='bold')
plt.ylabel('Frecuencia dias', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

plt.legend(
    title='Dirección Cardinal',
    title_fontsize=12,
    fontsize=10,
    loc='upper left',          
    bbox_to_anchor=(1.05, 1),   # Mover fuera del área del gráfico
    borderaxespad=0
)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
#UNSTACKED

plt.style.use('tableau-colorblind10')

crosstab_dir_viento_estaciones.plot(
    kind='bar', 
    stacked=False, 
    figsize=(10, 6), 
    color=sns.color_palette("mako", n_colors=len(crosstab_dir_viento_estaciones.columns))  
)

plt.title('Distribución de Direcciones Cardinales por Estación', fontsize=14, fontweight='bold', color='black')
plt.xlabel('Nombre de la Estación', fontsize=12, fontweight='bold')
plt.ylabel('Frecuencia dias', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)

plt.legend(
    title='Dirección Cardinal',
    title_fontsize=12,
    fontsize=10,
    loc='upper left',          
    bbox_to_anchor=(1.05, 1),   # Mover fuera del área del gráfico
    borderaxespad=0
)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()

plt.show()
```

```python
chi2, p, dof, expected = chi2_contingency(crosstab_dir_viento_estaciones)
print(f"Valor p: {p}")
```

Conclusion no hay relacion entre las estaciones y la direccion del viento

```python
## PEARSON

contaminantes = ['monoxido_nitrogeno', 'dioxido_nitrogeno', 'oxido_nitrogeno', 'ozono']
meteo = ['temperatura', 'vel_Viento', 'precipitacion']

correlaciones = {}

for meteo_var in meteo:
    for contaminante in contaminantes:
        corr_value = round(df_all_targets[meteo_var].corr(df_all_targets[contaminante]),3)
        correlaciones[f'corre_{meteo_var}_{contaminante}'] = corr_value

correlaciones
```

```python


df_all_targets_num = df_all_targets.select_dtypes('number')

correlation_matrix = df_all_targets_num.corr()

# Correlation Matrix-Heatmap Plot
mask = np.zeros_like(correlation_matrix)
mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

f, ax = plt.subplots(figsize=(20, 10))
sns.set(font_scale=1.25) # increase font size

ax = sns.heatmap(correlation_matrix, mask=mask, annot=True, annot_kws={"size": 10}, linewidths=.5, cmap="coolwarm", fmt=".2f", ax=ax) # round to 2 decimal places
ax.set_title("Dealing with Multicollinearity", fontsize=20) # add title
plt.show()
```

Interpretación de los Gráficos:
Monóxido de Nitrógeno (NO) vs. Temperatura: De acuerdo con la correlación -0.31, esperamos ver una relación débil y negativa, donde a medida que la temperatura aumenta, los niveles de monóxido de nitrógeno tienden a disminuir. Sin embargo, la relación no debería ser muy fuerte.

Dióxido de Nitrógeno (NO₂) vs. Temperatura: La correlación de -0.22 sugiere una relación débil y negativa, por lo que la dispersión de los puntos puede ser más amplia, y la línea de regresión no debería mostrar una tendencia clara.

Óxido de Nitrógeno (NOₓ) vs. Temperatura: Con una correlación de -0.29, la relación también será moderadamente negativa, similar al monóxido de nitrógeno. La línea de regresión puede ser levemente descendente.

Ozono (O₃) vs. Temperatura: Aquí la correlación es 0.63, lo que sugiere una relación positiva moderada. Se espera que la línea de regresión sea ascendente, indicando que, a medida que la temperatura aumenta, también lo hacen los niveles de ozono.

```python
### QUANTILE PLOTS (QQ PLOTS)

import statsmodels.api as sm

# Listas de contaminantes y variables meteorológicas
contaminantes = ['monoxido_nitrogeno', 'dioxido_nitrogeno', 'oxido_nitrogeno', 'ozono']
variables_meteo = ['temperatura', 'vel_Viento', 'precipitacion']

# Función para generar QQ plots
def generate_qq_plot(variable, title, color):
    plt.figure(figsize=(8, 6))
    sm.qqplot(variable, line='s', color=color, alpha=0.7)
    plt.title(f"QQ Plot - {title}", fontsize=16, fontweight='bold', color='darkblue')
    plt.xlabel('Cuantiles Teóricos', fontsize=12)
    plt.ylabel('Cuantiles de los Datos', fontsize=12)
    plt.grid(alpha=0.3, linestyle='--')
    plt.show()

# Generar QQ plots para contaminantes
for contaminante in contaminantes:
    generate_qq_plot(df_all_targets[contaminante].dropna(), f"Contaminante: {contaminante}", color="darkorange")

# Generar QQ plots para variables meteorológicas
for variable in variables_meteo:
    generate_qq_plot(df_all_targets[variable].dropna(), f"Variable Meteorológica: {variable}", color="teal")
```
